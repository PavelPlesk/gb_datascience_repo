```python
def kramer(A, b):
    x = np.zeros(b.shape)
    try:
        for i in np.arange(A.shape[1]):
            A_i = A.copy()
            A_i[:, i] = b
            x[i] = np.linalg.det(A_i) / np.linalg.det(A)
            print(f'x{i +1} = {x[i].round(3)}')
    except ZeroDivisionError:
        print('Ð¡Ð¸ÑÑ‚ÐµÐ¼Ð° ÑÐ²Ð»ÑÐµÑ‚ÑÑ Ð½ÐµÐ¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð½Ð¾Ð¹.')
        return
    return x
```

#### __4*.__ Ð ÐµÑˆÐ¸Ñ‚Ðµ ÑÐ¸ÑÑ‚ÐµÐ¼Ñƒ Ð»Ð¸Ð½ÐµÐ¹Ð½Ñ‹Ñ… ÑƒÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ð¹ Ð¼ÐµÑ‚Ð¾Ð´Ð¾Ð¼ Ð¥Ð¾Ð»ÐµÑ†ÐºÐ¾Ð³Ð¾:

$\begin{cases}
81x_{1}-45x_{2}+45x_{3}=531 \\

```python
def is_sim(A):
    if A.shape[0] == A.shape[1]:
        for i in np.arange(A.shape[0]):
            for j in np.arange(A.shape[0]):
                if A[i, j] != A[j, i]:
                    print('ÐÐµÐºÐ¾Ñ€Ñ€ÐµÐºÑ‚Ð½Ñ‹Ðµ Ð´Ð°Ð½Ð½Ñ‹Ðµ. ÐœÐ°Ñ‚Ñ€Ð¸Ñ†Ð° ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚Ð¾Ð² Ð´Ð¾Ð»Ð¶Ð½Ð° Ð±Ñ‹Ñ‚ÑŒ ÑÐ¸Ð¼Ð¼ÐµÑ‚Ñ€Ð¸Ñ‡Ð½Ð¾Ð¹.')
                    return False        
    else:
        print('ÐÐµÐºÐ¾Ñ€Ñ€ÐµÐºÑ‚Ð½Ñ‹Ðµ Ð´Ð°Ð½Ð½Ñ‹Ðµ. ÐœÐ°Ñ‚Ñ€Ð¸Ñ†Ð° ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚Ð¾Ð² Ð´Ð¾Ð»Ð¶Ð½Ð° Ð±Ñ‹Ñ‚ÑŒ ÐºÐ²Ð°Ð´Ñ€Ð°Ñ‚Ð½Ð¾Ð¹.')
        return False
    return True
```

```python
def holezky(A, b):
    if not is_sim(A):
        return
    l = np.zeros(A.shape)
    for j in np.arange(A.shape[0]):
        for i in np.arange(A.shape[0]):
            if i == j:
                l[i, j] = np.sqrt(A[i, i] - np.sum([l[i, k] ** 2 for k in np.arange(i)]))
            elif i > j:
                l[i, j] = (A[i, j] - np.sum([l[i, k] * l[j, k] for k in np.arange(i)])) / l[j, j]
    y = np.zeros(b.shape)   
    for i in np.arange(A.shape[0]):
        y[i] = (b[i] - np.sum([l[i, j] * y[j] for j in np.arange(i)])) / l[i, i]
    
    x = np.zeros(b.shape)
    for i in np.arange(A.shape[0] - 1, -1, -1):
        x[i] = (y[i] - np.sum([l.T[i, j] * x[j] for j in np.arange(i + 1, A.shape[0])])) / l.T[i, i]
        print(f'x{i+1} = {x[i].round(5)}')
    return x
```

```python
# ÐŸÑ€Ð¾Ð²ÐµÑ€ÐºÐ° Ð½Ð° Ð´Ð°Ð½Ð½Ñ‹Ñ… Ð¸Ð· Ð¼ÐµÑ‚Ð¾Ð´Ð¸Ñ‡ÐºÐ¸

A = np.array([[2, 1, 4], [1, 1, 3], [4, 3, 14]])
b = np.array([16, 12, 52])

holezky(A, b)
```
