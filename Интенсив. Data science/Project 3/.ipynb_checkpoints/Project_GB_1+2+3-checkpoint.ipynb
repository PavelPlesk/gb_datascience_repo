4. Проверить гипотезу о нормальности распределения для признаков 1 и 2.

```python
from scipy.stats import normaltest, shapiro, kstest, anderson, norm
```

```python
def normalize(sample):
    return (sample - sample.mean()) / sample.std(ddof=1)
```

```python
def anderson_pvalue(data):
    andersen = anderson(data)
    statistic = andersen.statistic
    critical_values = list(andersen.critical_values)
    significance_level = list(andersen.significance_level)
    for value, level in zip(critical_values, significance_level):
        if statistic < value:
            return level / 100
    return 0
```

```python
def pvalues(data):
    return pd.Series([normaltest(data)[1], 
                      shapiro(data).pvalue, 
                      kstest(normalize(data), 'norm').pvalue, 
                      anderson_pvalue(data)],
                      index=['normaltest', 'shapiro', 'kstest', 'anderson_pvalue'])
```

6. Написать функцию, автоматизирующую EDA.

```python
def calc_range(x):
    return x.max() - x.min()
```

```python
def describe(data, groupby=None):
    if groupby:
        desc = data.groupby(groupby).agg(['mean','std', ('var(n-1)', lambda x: x.var(ddof=1)),
                                  'min','max', ('range', lambda x: x.max() - x.min())]).T
    else:
        desc = data.agg(['mean', lambda x: x.var(ddof=1), 'min', 'max', calc_range]).T
        desc.columns = ['mean', 'var(n-1)', 'min', 'max', 'range']
    return desc
```

```python
def visualization(data, groupby=None):
    fig = plt.figure(figsize=(12, 7))
    fig.suptitle('Классификация признаков')
    sns.pairplot(data, hue=groupby)
    plt.show()
```

```python
def corr_heatmap(data):
    mask = np.tril(data.corr())
    sns.heatmap(data.corr(), annot=True, vmax=1, vmin=-1, center=0, cmap='RdYlGn', mask=mask)
    plt.show()
```

```python
def ntests(data, groupby=None, attrs="all"):
    classes = data[groupby].unique() if groupby else []
    attrs = list(data.columns) if attrs == "all" else attrs
    try:
        attrs.remove(groupby)
    except:
        pass
    for attr in attrs:
        if groupby:
            for cls in classes:
                mask = data[groupby] == cls  
                print(f'Normality checking for "{attr}" class "{cls}"', end='\n\n')
                try:
                    print(pvalues(data[mask][attr]), end='\n\n')
                except Exception as err:
                    print(err)
        else:
            print(f'Normality checking for "{attr}"', end='\n\n')
            try:
                print(pvalues(data[attr]), end='\n\n')
            except Exception as err:
                print(err)
```

```python
def eda(data, groupby=None, info=True, desc=True, vis=True, corrh=True, norm_chk='all'):
    
    print(data.info(), end='\n\n') if info else print("You can show info via 'info=True'", end='\n\n')
    print(describe(data, groupby), end='\n\n') if desc else print("You can show description via 'desc=True'", end='\n\n')
    visualization(data, groupby) if visualization else print("You can show visualization via 'vis=True'", end='\n\n')
    corr_heatmap(data) if corrh else print("You can show correlation heatmap via 'corrh=True'", end='\n\n')   
    ntests(data, groupby='Class', attrs=norm_chk) if ntests else print("You can perform normality tests via 'norm_chk=[list of attrs]'", end='\n\n') 
```

```python
eda(iris_df, groupby="Class", info=True, desc=True, vis=True, corrh=True, norm_chk=["Sepal length", "Sepal width"])
```
