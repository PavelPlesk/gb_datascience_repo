#### Разведочный анализ (EDA – exploratory data analysis)

```python
pd.set_option('display.float_format', '{:.1f}'.format)
```

Для сревнения правильно использовать результаты измерений, полученные при одинаковой температуре либо пересчитанные к одинаковой темпертуре по формуле. 

Приведем все результаты в столбце "IR60" к температуре 20 С. Полученные данные будем использовать для дальнейшего анализа.

```python
df["IR60 at 20"] = df["IR60"] * (1.04 ** (df["Temp"] - 20))
df.head()
```

```python
df.describe()
```

Ошибки в данных о технических параметрах трансформаторов (Power, Voltage) были выявлены и исправлены на этапе отбора данных.

Анализ датасета указывает на то, что более 75% трансформаторов, диагноостированных за последние годы составляют, так называемые, трансформаторы низкого напряжения (до 11 кВ) и мощностью до 2500 кВА.

Полученные результаты измерения сопротивления изоляции выше 100 000 МОм вызывают некоторые сомнения. 

Выполним поиск выбросов.

```python
def find_outliers(samples):
    q1 = samples.quantile(0.25)
    q3 = samples.quantile(0.75)
    iqr = q3 - q1
    outliers = samples[(samples < (q1 - 1.5 * iqr)) | (samples > (q3 + 1.5 * iqr))]
    return outliers
```

```python
outliers = pd.Series(dtype=float)
temp_df = df
i = 1
while True:
    new_outliers = find_outliers(temp_df["IR60 at 20"])
    outliers = pd.concat([outliers, new_outliers], ignore_index=True)
    print(i, " " * 5 ,new_outliers.count())
    temp_df = temp_df[~temp_df["IR60 at 20"].isin(outliers)]
    i += 1
    if new_outliers.count() == 0:
        break
        
print("\n", outliers.count(), f'{outliers.count() / df["IR60 at 20"].count() * 100:.1f} %')
```

На выбросы приходится более 51% данных.

```python
temp_df.describe()
```

Проверим на выбросы каждую из групп.

```python
mask = df["Voltage"] == 0.4
outliers_gr = pd.Series(df.loc[mask, "IR60 at 20"], dtype=float)
temp_df = df
temp_df = temp_df[~temp_df["IR60 at 20"].isin(outliers_gr)]
for i in range(1, 10):
    print(i)
    for group in set(temp_df["Voltage_group"]):
        mask = temp_df["Voltage_group"] == group
        new_outliers = find_outliers(temp_df.loc[mask, "IR60 at 20"])
        outliers_gr = pd.concat([outliers_gr, new_outliers], ignore_index=True)
        print(group, " " * (15 - len(group)) ,new_outliers.count())
    print("\n")
    temp_df = temp_df[~temp_df["IR60 at 20"].isin(outliers_gr)]

print("\n", outliers_gr.count(), f'{outliers_gr.count() / df["IR60 at 20"].count() * 100:.1f} %')
```

Количество потерянной информации уменьшилось до 31%.
