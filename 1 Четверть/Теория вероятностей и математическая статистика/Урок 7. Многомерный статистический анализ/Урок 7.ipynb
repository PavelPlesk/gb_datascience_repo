Подберите два признака (из четырёх) так, чтобы уровень линейной зависимости целевой переменной от значений этих признаков был максимальным. Другими словами, модель линейной регрессии на этих признаках должна давать наилучший результат.

```python
ones = np.ones((df.shape[0], 1))
X = np.hstack((ones, df[[f'X{j}', f'X{k}']].values))
X
```

```python
from itertools import combinations

res_df = pd.DataFrame()
y = df['y'].values
ones = np.ones((df.shape[0], 1))

m_count = df.filter(like='X', axis=1).shape[1]

for j,k in combinations(range(1, m_count + 1), 2):
    X = np.hstack((ones, df[[f'X{j}', f'X{k}']].values))

    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)

    det = np.linalg.det(XTX)

    b = XTX_inv.dot(X.T).dot(y)

    z = X.dot(b)

    R2 = 1 - ((y - z)**2).sum() / ((y - y.mean()) ** 2).sum()

    data = {'vars':f'X{j}, X{k}'}
    data.update({f'b{l}':b[l] for l in range(0, b.shape[0])})
    data.update({'det':det})
    data.update({'R2':R2})

    res_df = res_df.append(data, ignore_index=True)
res_df
```

```python
opt = res_df.sort_values(by='R2', ascending=False).head(1)
opt
```

Наибольшая зависимость наблюдается от признаков X1 и X3.

Является ли значимым получившееся уравнение регрессии?

```python
n = df.shape[0]
m = 2

k1 = m
k2 = n - m - 1

alpha = 0.05

t = stats.f.ppf(1 - alpha, k1, k2)
t
```

```python
R2 = opt.iloc[0]['R2']
F = (R2 / k1) / ((1 - R2) / k2)
F
```

```python
F > t
```

Получившееся уравнение является значимым с уровнем значимости 5%. 

Выбранная пара признаков объясняет 76.3 % дисперсии переменной y.

#### Задача 2

Даны значения величины заработной платы заемщиков банка (salary) и значения их поведенческого кредитного скоринга (scoring):

salary = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110]
scoring = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]

Возьмём в качестве признака значение salary, а в качестве целевой переменной - scoring.

```python
salary = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110]) 
scoring = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

Найдите коэффициенты линейной регрессии с помощью формул для парной регрессии, а затем с помощью метода наименьших квадратов.

```python
res_df = pd.DataFrame()

x1 = salary
y1 = scoring

b1_1 = (np.mean(x1 * y1) - np.mean(x1) * np.mean(y1)) / (np.mean(x1 ** 2) - np.mean(x1) ** 2)
b0_1 = np.mean(y1) - b1_1 * np.mean(x1)

res_df = res_df.append({'method':'manual', 'b0':b0_1, 'b1':b1_1}, ignore_index=True)
```

```python
x2 = salary.reshape(-1,1)
y2 = scoring

ones = np.ones((x2.shape[0], 1))

x2 = np.hstack((ones, x2.reshape(-1,1)))

XTX = x2.T.dot(x2)
XTX_inv = np.linalg.inv(XTX)

det = np.linalg.det(XTX)

b = XTX_inv.dot(x2.T).dot(y2)

res_df = res_df.append({'method':'minq', 'b0':b[0], 'b1':b[1]}, ignore_index=True)

res_df
```

Постройте scatter plot по данным и отметьте на нём прямую линейной регрессии, полученную в п. 1.

```python
plt.scatter(salary, scoring)

ox = np.linspace(salary.min(), salary.max(), 2)
oy = b[0] + ox * b[1]
plt.plot(ox, oy, color='darkred')

plt.show()
```

Посчитайте коэффициент детерминации.

```python
z2 = x2.dot(b)

R2 = 1 - ((y2 - z2)**2).sum() / ((y2 - y2.mean()) ** 2).sum()
R2
```

Оцените построенное уравнение регрессии с помощью F-критерия Фишера.

```python
n = y2.shape[0]
m = 1

k1 = m
k2 = n - m - 1

alpha = 0.05

t = stats.f.ppf(1 - alpha, k1, k2)

F = (R2 / k1) / ((1 - R2) / k2)

F > t, F, t
```

Уравнение значимое для уровня значимости 5 % и описывает 78,7% дисперсии показателя scoring.
