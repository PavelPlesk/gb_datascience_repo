#### Задача 1

Известно, что генеральная совокупность распределена нормально с известным средним квадратическим отклонением 16. Найти доверительный интервал для оценки математического ожидания mu с надёжностью 0.95 (доверительная информация), если выборочное среднее равно 80, а объём равен 256.

```python
std = 16
p = 0.95
mean = 80
n = 256
```

```python
alpha = 1 - p

t1 = stats.norm.ppf(alpha / 2)
t2 = stats.norm.ppf(1 - alpha / 2)

t1, t2
```

Доверительный интервал

```python
(mean + t1 * std / np.sqrt(n), mean + t2 * std / np.sqrt(n))
```

```python
ox = np.linspace(40, 120, 500)
oy = stats.norm.pdf(ox, loc=80, scale=16)
plt.plot(ox, oy)

ox_p95 = np.linspace(mean + t1 * std / np.sqrt(n), mean + t2 * std / np.sqrt(n), 100)
oy_p95 = stats.norm.pdf(ox_p95, loc=80, scale=16)

plt.fill_between(ox_p95, oy_p95, alpha=0.5, color='C0')
```

#### Задача 2

Продавец утверждает, что средний вес пачки печенья составляет 200 г. Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:

202, 203, 199, 197, 195, 201, 200, 204, 194, 190

Известно, что их веса распределены нормально.
1. Верно ли утверждение продавца, если учитывать, что уровень значимости равен 1%?
2. Найдите P-значение для данного теста.

Н0: mu = 200

H1: mu != 200

```python
mu = 200
alpha = 0.01
samples = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
n = samples.size
```

Используем t-статистику для данных с неизвестной дисперсией

```python
def t_statistic(samples, mu):
    return (samples.mean() - mu) / samples.std(ddof=1) * np.sqrt(samples.size - 1)
```

```python
t1 = stats.t.ppf(alpha / 2, df = n - 1)
t2 = stats.t.ppf(1 - alpha / 2, df = n - 1)

t1, t2
```

```python
t = t_statistic(samples, mu)
t
```

```python
ox = np.linspace(t1 * 1.5, t2 * 1.5, 500)
oy = stats.t.pdf(ox, df=n-1)

ox_left = np.linspace(t1 * 1.5, t1, 100)
oy_left = stats.t.pdf(ox_left, df=n-1)

ox_right = np.linspace(t2, t2 * 1.5, 100)
oy_right = stats.t.pdf(ox_right, df=n-1)

plt.plot(ox, oy)
plt.fill_between(ox_left, oy_left, alpha=0.5, color='C0')
plt.fill_between(ox_right, oy_right, alpha=0.5, color='C0')

plt.axvline(x=t, ymin=0.04, ymax=0.96, color='red')
```

Утверждение продавца верно

P-value для двухсторонней области

```python
S = t_statistic(samples, mu)

p_left = stats.t.cdf(S, df=n-1)
p_right = 1 - stats.t.cdf(S, df=n-1)

pvalue = 2 * min(p_left, p_right)

pvalue
```

```python
pvalue > alpha
```

#### Задача 3

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм. Используя уровень значимости 0.025 и правостороннюю критическую область, проверить данную гипотезу при условии, что в выборке из 100 шариков средний диаметр оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.

H0: mu = 17

H1: mu != 17

```python
mu = 17
alpha = 0.025
n = 100
mean = 17.5
std = 2
```

Поскольку дисперсия известна используем Z-статистику

```python
def z_statistic(mu):
    return (mean - mu) / std * np.sqrt(n - 1)
```

С учетом того, что критическая область односторонняя (правая)

```python
t2 = stats.norm.ppf(1 - alpha)
t2
```

```python
t = z_statistic(mu)
t
```

```python
ox = np.linspace(t2 * (-1.5), t2 * 1.5, 500)
oy = stats.norm.pdf(ox)

ox_right = np.linspace(t2, t2 * 1.5, 100)
oy_right = stats.norm.pdf(ox_right)

plt.plot(ox, oy)
plt.fill_between(ox_right, oy_right, alpha=0.5, color='C0')

plt.axvline(x=t, ymin=0.04, ymax=0.96, color='red')
```

P-value для правосторонней критической области

```python
S = z_statistic(mu)

pvalue = 1 - stats.norm.cdf(S)
pvalue
```

```python
pvalue > alpha
```

Гипотеза не верна

#### Задача 4 (Дополнительно)

Дана выборка диаметров подшипников из примера 1 с занятия 5:

samples = [0.6603, 0.9466, 0.5968, 1.3792, 1.5481, 0.7515, 1.0681, 1.1134,
1.2088, 1.701 , 1.0282, 1.3579, 1.0191, 1.1784, 1.1168, 1.1372,
0.7273, 1.3958, 0.8665, 1.5112, 1.161 , 1.0232, 1.0865, 1.02 ]

Предполагая, что диаметры подшипников распределены нормально, проверьте гипотезу о том, что дисперсия случайной величины равна 0.0625 при уровне значимости alpha = 0.05. Что для этого нужно знать:
1. Альтернативная гипотеза двухсторонняя.
2. Статистика для теста: H = (n - 1) * sample_variance / variance, где n - число элементов в выборке, sample_variance - несмещённая оценка дисперсии, variance - утверждаемая нулевой гипотезой дисперсия.
3. Эта статистика в предположении верности нулевой гипотезы имеет распределение хи-квадрат с параметром df = n - 1. Её квантили можно найти с помощью функции scipy.stats.chi2.ppf.

H0: std = 0.25

H1: std != 0.25

```python
std = 0.25
samples = np.array([0.6603, 0.9466, 0.5968, 1.3792, 1.5481, 0.7515, 1.0681, 1.1134, 1.2088, 1.701 , 1.0282, 1.3579, 1.0191, 1.1784, 1.1168, 1.1372, 0.7273, 1.3958, 0.8665, 1.5112, 1.161 , 1.0232, 1.0865, 1.02 ])
alpha = 0.05
```

Для проверки гипотезы используем статистику Хи2, распределенную по Пирсону с числом степеней свободы n-1

```python
def xi2(samples, std):
    return (samples.size - 1) * samples.std(ddof=1) ** 2 / std ** 2
```

```python
n = samples.size

t1 = stats.chi2.ppf(alpha / 2, df=n-1)
t2 = stats.chi2.ppf(1 - alpha / 2, df=n-1)

t1, t2
```

```python
t = xi2(samples, std)
t
```

```python
ox = np.linspace(t1 * 0.8, t2 * 1.2, 500)
oy = stats.chi2.pdf(ox, df=n-1)

ox_left = np.linspace(t1 * 0.8, t1, 100)
oy_left = stats.chi2.pdf(ox_left, df=n-1)

ox_right = np.linspace(t2, t2 * 1.2, 100)
oy_right = stats.chi2.pdf(ox_right, df=n-1)

plt.plot(ox, oy)
plt.fill_between(ox_left, oy_left, alpha=0.5, color='C0')
plt.fill_between(ox_right, oy_right, alpha=0.5, color='C0')

plt.axvline(x=t, ymin=0.04, ymax=0.96, color='red')
```

```python
H = xi2(samples, std)

p_left = stats.chi2.cdf(H, df=n-1)
p_right = 1 - stats.chi2.cdf(H, df=n-1)

pvalue = 2 * min(p_left, p_right)

pvalue
```

```python
pvalue > alpha
```

Гипотеза верна
